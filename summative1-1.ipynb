### Cumulative Line Graph of Attendance Rate Over Time

#### Reasoning for an Additional Cumulative (Weighted) Attendance Rate  

* This graph provides a long-term view of attendance trends by showing how overall participation changes cumulatively over time, rather than focusing on individual sessions 
* Using a weighted cumulative rate ensures that each class session contributes proportionally to the overall trend - larger classes have a greater impact than smaller ones, preventing skewed results  
* The x-axis represents the date of each class session for the Algorithms module, arranged chronologically from early 2023 to late 2025 
* The y-axis represents the cumulative attendance rate (% that attended out of total students invited) up to each point in time, accounting for all previous sessions
* This approach smooths out short-term fluctuations seen in daily averages, revealing the underlying pattern of student engagement across multiple years 
* To create this plot, I will:  
  1. Filter the dataset to include only the Algorithms module
  2. Calculate the total number of students invited and attended for each date
  3. Use cumulative sums of attendance and invitations to compute the overall attendance rate over time  
  4. Plot the weighted cumulative attendance as a line graph to visualise sustained participation trends  
* Overall, a cumulative view provides a more well rounded measure of module attendance than daily attendance rates alone

````python
# Daily number invited (row count per date)
invited_per_date = (algorithms_attendance.groupby("Date")["Person Code"].count().reset_index(name="Invited_Count"))

# Daily number attended (sum of 1/0 per date)
attended_per_date = (algorithms_attendance.groupby("Date")["Attended"].sum().reset_index(name="Attended_Count"))

# Merge daily numbers invited/attended to get a single daily table
daily_counts = pd.merge(invited_per_date, attended_per_date, on="Date", how="inner").sort_values("Date")
````

````python
# Print table to check
# "Attended_Count" should always be less than or equal to "Invited_Count"
print(daily_counts.head())
````

````python
# Cumulative sums of counts
daily_counts["Cum_Attended"] = daily_counts["Attended_Count"].cumsum()
daily_counts["Cum_Invited"]  = daily_counts["Invited_Count"].cumsum()

# Weighted cumulative rate (can never exceed 100%)
daily_counts["Weighted_Cum_Rate"] = (daily_counts["Cum_Attended"] / daily_counts["Cum_Invited"]) * 100
````

````python
# Print cumulative totals to check
print(daily_counts[["Date", "Cum_Attended", "Cum_Invited", "Weighted_Cum_Rate"]].head())
print("Check: Weighted_Cum_Rate should increase gradually and never exceed 100%")
````

````python
# Cumulative line graph
plt.plot(daily_counts["Date"], daily_counts["Weighted_Cum_Rate"], marker="o", color="steelblue")
plt.title("Cumulative (Weighted) Attendance Rate (Algorithms Module)")
plt.xlabel("Date")
plt.ylabel("Cumulative Attendance Rate (% of total invited)")
plt.xticks(rotation=45) # rotate x-axis labels for better readability
plt.grid(True) # gridlines for readability

plt.tight_layout()
plt.show()
````

#### Description
* The graph shows the cumulative (weighted) attendance rate for the Algorithms module from early 2023 to late 2025
* Overall attendance remains relatively stable, with cumulative participation fluctuating between around 65% and 85% over the two-year period
* The rate begins high in early 2023, reflecting strong initial engagement, but steeply declines in early 2023, then continues to gradually decline through late 2023 and early 2024
* A slow upward trend begins mid-2024, suggesting improved attendance consistency or higher turnout in later sessions
* The overall shape indicates that later sessions had slightly higher attendance rates, gradually lifting the cumulative average after an earlier dip
* The cumulative nature of the graph smooths short-term fluctuations, providing a clearer view of long-term participation trends
* The curveâ€™s flattening toward the end of 2025 implies that attendance rates were becoming more consistent and steady across sessions

#### Comparison of Cumulative vs. Regular Attendance Graphs  

* The regular line and bar graphs show average attendance for each individual session, highlighting short-term fluctuations across different dates 
* Attendance in these graphs is highly variable, with several sharp rises and drops â€” including multiple peaks near 100% and dips below 50%, especially in early 2023, early 2024, and mid-2025  
* These rapid changes reflect session-by-session variability, such as differences in class size, scheduling, or student engagement
* In contrast, the cumulative graph smooths out these short-term swings by showing the overall attendance trend over time, weighted by class size 
* The cumulative line starts high, dips slightly through late 2023 and early 2024, and then gradually increases again from mid-2024 onwards, indicating a slow recovery in long-term attendance consistency  
* Unlike the regular graphs, the cumulative curve cannot exceed 100%, and its shape reflects aggregate participation rather than daily spikes  
* Together, the graphs provide complementary insights:  
  - The **line and bar graphs** reveal when attendance fluctuated most 
  - The **cumulative graph** captures the broader trend in sustained engagement across the full module period

````python
# 3) compute the Pearson correlation coefficient
    # Now we actually run the correlation test using pearsonr(x, y)

r_pearson, p_pearson = pearsonr(x, y)

# Print title for results
print("Pearson correlation between Module Mean Att. and Mean Overall Att. of Students:")

print('correlation:', r_pearson)
print('p-value:', p_pearson)
print('r squared:', r_pearson**2)

# This will return 3 values:
    # r_pearson: correlation coefficient (strength and direction of linear relationship)
    # p_pearson: p-value (evidence against the null hypothesis of no correlation)
    # r_pearson**2: proportion of variance explained by the linear relationship
````

````python
# 4) Additional correlation matrix using pandas
    # This is just to check the correlation coefficient of the above model
        # pandas .corr() calculates the correlation between all number columns
        # It doesn't provide a p-value meaning that we are reliant on the pearsonr model for significance

corr_matrix = module_clean[["Module Mean Att.", "Mean Overall Att. of Students"]].corr()

print("Correlation matrix from pandas .corr():")
print(corr_matrix)
````

**Key Findings from the Correlation Analysis**
* There is a strong positive correlation between:
    * Module Mean Attendance (how well-attended a module is overall), and
    * Mean Overall Attendance of Students who attend that module.
* The Pearson correlation coefficient is r = 0.683, indicating a substantial, meaningful relationship.
    * The p-value is < 0.001, showing the correlation is highly statistically significant â€” extremely unlikely to be due to chance.

**Interpretation:**
* This correlation indicates that modules with higher attendance tend to be populated by students who also have higher overall attendance, whereas lower-attendance modules tend to have students with lower overall attendance
* This provides strong evidence that attendance behaviour clusters, with high-attending students gravitate toward (or remain consistent in) well-attended modules, and low-attenders cluster in lower-attendance modules.
* However, it is important to note that correlation does not imply causation, meaning this analysis does not demonstrate that:
    * High-attendance students deliberately choose high-attendance modules
    * High module attendance causes students to attend more
    * Low-attendance modules attract low-attendance students

**The correlation only indicates an association, not a causal mechanism**

### Linear Regression

**Steps for linear regression analysis:**
1) I already have the module_corr dataframe from the correlation analysis, which contains the two variables needed for regression:
    * Module Mean Att. = predictor (independent variable)
    * Mean Overall Att. of Students = outcome (dependent variable)
2) I need to set up a regression formula using statsmodels:
    * The formula will take the form: "Mean Overall Att. of Students" ~ "Module Mean Att."
3) I can then fit the linear regression model using statsmodels.formula.ols()
    * This estimates how module attendance predicts the average overall attendance of students who attend that module
4) I can then extract key statistics from my regression model for my analysis
5) I then will be able to print and interpret the results of the linear regression analysis

>**Interpretation logic:**
>* If the coefficient (coef) for Module Mean Att. is positive and statistically significant (P>|t| < 0.05) = modules with higher mean attendance tend to have students with higher overall attendance.
>* If the coefficient (coef) for Module Mean Att. is negative and statistically significant (P>|t| < 0.05) = modules with lower mean attendance tend to have students with lower overall attendance.
>* If the coefficient for Module Mean Att. is close to zero and/or not significant (P>|t| â‰¥ 0.05) = module attendance does not meaningfully predict studentsâ€™ overall attendance.
>* If R-squared is reasonably large (e.g. > 0.25) = module attendance explains a substantial proportion of the variation in student overall attendance.
>* If R-squared is small (e.g. < 0.10) = very little of the variation in student overall attendance is explained by module attendance, even if the coefficient is statistically significant.
>* If the F-statistic is significant (Prob(F-statistic) < 0.05) = the regression model as a whole provides a statistically significant prediction of student overall attendance.

````python
# 2) Set up the regression formula

# Import statsmodels.formula.api
import statsmodels.formula.api as smf

# Define the formula
formula = 'Q("Mean Overall Att. of Students") ~ Q("Module Mean Att.")'
    # Because our column names contain spaces and punctuation, we wrap them with Q("...")

# Print the formula to check
print("Regression formula being used:")
print(formula)
````

````python
# 3) Fit the linear regression model
    # We use statsmodels' ols() function with the formula and the module_corr dataframe.

model = smf.ols(formula=formula, data=module_clean)

# Fit the model to the data
model_fitted = model.fit()

# Print the summary of the regression results
print(model_fitted.summary())
````

````python
# 4) Additionally, print key stats

# Get the full table of coefficients
coef_table = model_fitted.params
pvalues = model_fitted.pvalues
r_squared = model_fitted.rsquared

print("Coefficient estimates (coef):")
print(coef_table)

print("P-values for each coefficient (P>|t|):")
print(pvalues)

print("R-squared:")
print(r_squared)
````

**Visualisations for regression model**

````python
# Regression plot with fitted line
sns.regplot(
    data=module_clean,
    x="Module Mean Att.",
    y="Mean Overall Att. of Students",
)

plt.title("Regression Plot")
plt.xlabel("Module Mean Attendance")
plt.ylabel("Mean Overall Attendance of Students")

plt.show()
````

````python
# Residuals plots

# Create the subplot figure
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1) Residuals vs fitted
    # This plot helps us see whether residuals are randomly scattered (good) or if there is a pattern (which might suggest non-linearity)
plt.subplot(1, 3, 1)
sns.residplot(
    data = module_clean,
    x = "Module Mean Att.",                 # predictor variable
    y = "Mean Overall Att. of Students",    # outcome variable
    color = "red"
)

plt.xlabel("Module Mean Attendance", fontsize=12)
plt.ylabel("Residuals", fontsize=12)
plt.title("Residuals vs Module Mean Attendance")

# 2) Histogram of residuals
    # This plot shows the distribution of the residuals, helping us assess whether they are roughly symmetric and approximately normal (a key model assumption)

# Extract observed values (y) and predicted values (y_pred)
y = module_clean["Mean Overall Att. of Students"]
y_pred = model_fitted.fittedvalues   # predicted values from the regression model

# Compute residuals
residuals = y - y_pred

# Plot histogram of residuals
plt.subplot(1, 3, 2)
plt.hist(residuals, bins=8)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Histogram of Residuals")

# 3) Q-Q plot of residuals
    # This plot compares the distribution of residuals to a theoretical normal distribution.
    # If residuals are roughly normal, the points should fall close to the straight line.

# Import probplot from scipy.stats
from scipy.stats import probplot

# Plot Q-Q plot
plt.subplot(1, 3, 3)
probplot(residuals, plot=plt)
plt.xlabel("Theoretical quantiles")
plt.ylabel("Observed residual quantiles")
plt.title("Q-Q Plot of Residuals")

plt.tight_layout()
plt.show()
````

**Key Findings from the Linear Regression Analysis**
* A simple linear regression was conducted to examine whether Module Mean Attendance predicts the Mean Overall Attendance of Students who attend each module
* The model shows a strong, statistically significant positive relationship between the two variables:
	* The slope coefficient for Module Mean Att. is 0.4713, indicating that modules with higher average attendance tend to have students with higher overall attendance
	* The p-value for this coefficient is < 0.001, showing that the relationship is statistically significant
* The regression visualisation clearly shows an upward trend, with a fitted line that closely reflects the positive association observed in the numerical output
* The modelâ€™s R-squared value is 0.466, meaning that 46.6% of the variation in studentsâ€™ overall attendance is explained by differences in module attendance
	* This represents a substantial proportion of explained variance for a model with a single predictor
* Residual diagnostics (residuals plot, histogram, and Qâ€“Q plot) suggest that model assumptions are reasonably met:
	* Residuals appear randomly scattered with no strong pattern (supporting linearity and homoscedasticity)
	* The histogram of residuals shows an approximately symmetric distribution
	* The Qâ€“Q plot indicates that residuals follow the normality assumption closely, with only small deviations at the tails

**Interpretation:**
* The regression results indicate that higher-attendance modules are associated with students who also have higher overall attendance, and lower-attendance modules are associated with students who have lower overall attendance
* The diagnostic plots reinforce that this linear relationship is statistically meaningful and not driven by major violations of model assumptions
* This suggests that attendance behaviour clusters within modulesâ€”students with similar attendance patterns tend to appear in modules with similar attendance characteristics
* However, it is important to note that regression does not imply causation, meaning the model does not demonstrate that:
	* Students choose high-attendance modules because of their own attendance patterns
	* Module attendance causes students to attend more (or less)
	* Low-attendance modules specifically attract lower-attendance students

**The regression shows prediction and association, not a causal mechanism**
